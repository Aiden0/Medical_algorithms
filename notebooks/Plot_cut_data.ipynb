```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
import time
```

```python
cut_ppg_times = np.load("../ppg_times.npy")
#RAW PPG date
cut_ppg_dat = np.load("../ppg_dat.npy")
cut_hrv_times = np.load("../hrv_times.npy")
cut_hrv_dat = np.load("../hrv_dat.npy")

```

```python
fati = np.load("../data/fatig.npy")
```

```python
%matplotlib notebook
ax = plt.subplot(2,1,1)
for i in range(len(cut_ppg_times)):
    plt.plot(cut_ppg_times[i],cut_ppg_dat[i])
plt.gcf().autofmt_xdate()
plt.setp(ax.get_xticklabels(), fontsize=6)

plt.subplot(2,1,2, sharex = ax)
for i in range(len(cut_hrv_times)):
    plt.plot(cut_hrv_times[i],cut_hrv_dat[i])

for i in range(len(fati)):
    plt.plot([fati[i], fati[i]],[0,100], 'k')
    
plt.gcf().autofmt_xdate()
```
